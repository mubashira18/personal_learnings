# working with mathematical formulaas

```python
a = np.arange(10)
np.sin(a)
```

```python
#sigmoid
def sigmoid(array):
    return 1/(1+np.exp(-array))

a = np.arange(10)
sigmoid(a)
```

```python
#mean squared error

actual = np.random.randint(1,50,25)
predicted = np.random.randint(1,50,25)

def mse(actual,predicted):
    return np.mean((actual-predicted)**(3-1))

mse(actual,predicted)
```
